```python
import pandas as pd
import seaborn as sns
import numpy as np
german = pd.read_csv("data/german_credit_data.csv")
```

```python
cols = german.columns[:] 
# определяем цвета 
# красный - пропущенные данные
colours = ['#eeeeee', '#ff0000']
sns.heatmap(german[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
german.isnull().any() 
```

```python
for col in german.columns:
    pct_missing = german[col].isnull().sum()
    print('{} - {}'.format(col, round(pct_missing)))
```

```python
german=german.drop(["Checking account"],axis=1)
german.head()
```

```python
german["Saving accounts"]=german["Saving accounts"].fillna("quite little")
german["Saving accounts"]
```

```python
for col in german.columns:
    pct_missing = german[col].isnull().sum()
    print('{} - {}'.format(col, round(pct_missing)))
```

```python
sns.histplot(data=german["Saving accounts"], log_scale=(False, True), bins=30)
```

```python
start_german = pd.read_csv('data\german_credit_data.csv')
sns.histplot(data=start_german["Saving accounts"], log_scale=(False, True), bins=30)
```

```python
# функция map() используется для применения функции к каждому элементу итерируемого объекта
german["Saving accounts"]=german["Saving accounts"].map({"quite little": 0, "little": 1, "moderate": 2,"rich": 3, "quite rich": 4})
german.head()
```

```python
german["Sex"]=german["Sex"].map({"male": 1, "female": 0})
german["Housing"]=german["Housing"].map({"own": 0, "free": 1, "rent": 2})
german.head()
```

```python
german['Age'].describe()
```

```python
print(np.where(german['Age']>65))
```

```python
german = german.apply(lambda x: x[(x>german.loc[0,x.name]) & 
                                    (x < german.loc[3,x.name])], axis=0)
```

```python
german = pd.DataFrame(german)

german.to_csv('data/german_credit_data_clear.csv', index=False)
```
